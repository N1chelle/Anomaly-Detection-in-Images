### _Calculating Structural Similarity Index for images in the test folder and finding the F-1 score_

## Importing libraries

```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
%matplotlib inline  
import glob
from skimage.metrics import structural_similarity
import argparse
import imutils
import cv2
```

## Creating Arrays of image pixels

```python
def list_img(num):
    num = str(num)
    path_num = "C:\\Users\\Nichelle Machado\\Becode Projects\\becode_dice_data\\train_set\\"+num
    list_num = []
    for filename in glob.iglob(f'{path_num}/*'):
        img = cv2.imread(filename, 0)
        list_num.append(img)
    return list_num
```

```python
array_00 = np.array(list_img('00'))
array_01 = np.array(list_img('01'))
array_02 = np.array(list_img('02'))
array_03 = np.array(list_img('03'))
array_04 = np.array(list_img('04'))
array_05 = np.array(list_img('05'))
array_06 = np.array(list_img('06'))
array_07 = np.array(list_img('07'))
array_08 = np.array(list_img('08'))
array_09 = np.array(list_img('09'))
array_10 = np.array(list_img('10'))
```

## Creating Templates for each dice face

```python
def template(num):
    num = str(num)
    t = np.mean(list_img(num), axis=0)
    return t
```

```python
temp= {}
temp['00'] = template('00')
temp['01'] = template('01')
temp['02']= template('02')
temp['03'] = template('03')
temp['04'] = template('04')
temp['05'] = template('05')
temp['06'] = template('06')
temp['07'] = template('07')
temp['08'] = template('08')
temp['09'] = template('09')
temp['10'] = template('10')
```

## Calculating SSIM for each image against each template

```python
def ssim(num1, num2):
    path = "C:\\Users\\Nichelle Machado\\Becode Projects\\becode_dice_data\\test_set\\"+num1
    list_ssim = []
    for filename in glob.iglob(f'{path}/*'):
        img = cv2.imread(filename, 0)
        (score, diff) = structural_similarity(img, temp[num2], full=True)
        diff = (diff * 255).astype("uint8")
        list_ssim.append(score)
    return list_ssim
```

```python
True_Positives = 138
False_Negatives = 82
False_Positives = 11+15
True_negatives = 37
```
